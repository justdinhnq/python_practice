```python
import matplotlib.pyplot as plt
import numpy as np

# the IPL batting average scores of the last 10 seasons (2010-2019) of Dhoni
X = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019])
dhoni = np.array([31.88, 43.55, 29.83, 41.90, 74.20, 31.00, 40.57, 26.36, 75.83, 83.20])
```

**Problem:** could you make anything out of this data? OR Were you able to extract any information out of the data?

**Step 1** - plots

```python
# one row, 2 column
fig, axis = plt.subplots(1, 2, figsize=(13, 6))

# scatter plot
axis[0].scatter(X, dhoni)

# line plot
axis[1].plot(X, dhoni)

# x_line
# from 'bottom left': move right by 0.5, move up by 0.04
# ha: horizontal align
fig.text(
    0.5, 0.04, 'Year', 
    ha = 'center', fontsize = 18
)

# y_line
fig.text(
    0.04, 0.5, 'Average Scores in IPL Seasons', 
    va='center', rotation='vertical', fontsize=18
)
```

We can't extract any useful information out of these plots. 

Let's add a trend line to a line chart.

```python
fig, axis = plt.subplots(1, 1, figsize = (10, 8))

# Least squares polynomial fit
# return: 
# a vector of coefficients p that minimises the squared error 
# in the order deg, deg-1,  0.
z = np.polyfit(x = X, y = dhoni, deg = 1)

# one-dimensional poly
# poly1d([1, 2, 3]) returns (x-1)(x-2)(x-3)
p = np.poly1d(z)

# p(X): evaluate the polynomial at x = X
plt.plot(X, p(X), "r--")                     # a trend line

plt.plot(X, dhoni)
plt.title('Dhoni Batting Average Scores', fontsize = 16)
plt.xlabel('Years', fontsize = 16)
plt.ylabel('Average Scores in Seacons', fontsize = 16)
```

Now we can extract information from this plot!

The plot is shown to have an upward trend representing that Dhoni looks to be playing well

**Storytelling - communicating story**

- the visualization is ready, it's very important part to communicate the story in a manner that actionable insights could be derived. 
- The ultimate goal is to help decision-makers take the decision

**The story:**

Dhoni looks to be playing well! His batting average is seen to have an upward trend which means that he can be trusted to play well in the upcoming season. Thus, an informed decision can be taken to invest in him.
